## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    # Variables para almacenar los totales
    total_importe = 0
    total_cantidad = 0
    
    # Itera sobre cada venta en la lista de datos
    for venta in datos:
        # Calcula el importe de la venta actual
        importe = venta['precio'] * venta['cantidad']
        
        # Suma el importe y la cantidad al total
        total_importe += importe
        total_cantidad += venta['cantidad']
    
    # Retorna los totales calculados
    return total_importe, total_cantidad

# Llama a la función calcular_totales con los datos cargados
importe, cantidad = calcular_totales(datos)

# Muestra los resultados
print(f"Las ventas fueron de ${importe:,.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
    # Diccionario para almacenar la cantidad total vendida de cada producto
    total_por_producto = {}
    
    # Itera sobre cada venta en la lista de datos
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        # Si el producto ya está en el diccionario, suma la cantidad
        if producto in total_por_producto:
            total_por_producto[producto] += cantidad
        # Si el producto no está en el diccionario, lo añade con la cantidad actual
        else:
            total_por_producto[producto] = cantidad
    
    # Retorna el diccionario con las unidades vendidas por producto
    return total_por_producto

def listar_ventas(ventas):
    # Imprime los productos y sus unidades vendidas
    print("Unidades vendidas por producto:")
    for producto, cantidad in ventas.items():
        print(f"{producto}: {cantidad} unidades")

# Llama a la función unidades_vendidas con los datos cargados
ventas = unidades_vendidas(datos)

# Llama a la función listar_ventas para mostrar las unidades vendidas por producto
listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio(datos):
    # Diccionario para almacenar la suma de precios y la cantidad de ventas por producto
    suma_precios = {}
    cantidad_ventas = {}
    
    # Itera sobre cada venta en la lista de datos
    for venta in datos:
        producto = venta['producto']
        precio = venta['precio']
        
        # Suma los precios y cuenta las ventas por producto
        if producto in suma_precios:
            suma_precios[producto] += precio
            cantidad_ventas[producto] += 1
        else:
            suma_precios[producto] = precio
            cantidad_ventas[producto] = 1
    
    # Calcula el precio promedio por producto
    precios_promedio = {}
    for producto in suma_precios:
        precios_promedio[producto] = suma_precios[producto] / cantidad_ventas[producto]
    
    # Retorna el diccionario con el precio promedio por producto
    return precios_promedio

def listar_precios(precios):
    # Imprime los productos y sus precios promedio
    print("Precio promedio por producto:")
    for producto, precio in precios.items():
        print(f"{producto}: ${precio:.2f}")

# Llama a la función precio_promedio con los datos cargados
precios = precio_promedio(datos)

# Llama a la función listar_precios para mostrar el precio promedio por producto
listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    # Diccionario para almacenar la cantidad total vendida de cada producto
    total_por_producto = {}
    
    # Itera sobre cada venta en la lista de datos
    for venta in datos:
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        # Si el producto ya está en el diccionario, suma la cantidad
        if producto in total_por_producto:
            total_por_producto[producto] += cantidad
        else:
            total_por_producto[producto] = cantidad
    
    # Ordena los productos por cantidad vendida de mayor a menor
    ranking = sorted(total_por_producto.items(), key=lambda x: x[1], reverse=True)
    
    # Retorna los primeros 'top' productos del ranking
    return ranking[:top]

def listar_ranking(ranking):
    # Imprime el ranking de productos
    print("Ranking de los productos más vendidos:")
    for i, (producto, cantidad) in enumerate(ranking, 1):
        print(f"{i}. {producto}: {cantidad} unidades vendidas")

# Llama a la función ranking_productos con los datos cargados y obtiene el top 3
ranking = ranking_productos(datos)

# Llama a la función listar_ranking para mostrar el ranking de productos
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from collections import defaultdict

def ventas_por_mes(datos):
    # Diccionario anidado para almacenar las ventas por mes y producto
    ventas_mensuales = defaultdict(lambda: defaultdict(int))
    
    # Itera sobre cada venta en la lista de datos
    for venta in datos:
        fecha = venta['fecha']
        producto = venta['producto']
        cantidad = venta['cantidad']
        
        # Extrae el mes y año de la fecha (formato 'YYYY-MM')
        mes = fecha[:7]
        
        # Suma la cantidad vendida al producto correspondiente en el mes específico
        ventas_mensuales[mes][producto] += cantidad
    
    # Convierte el defaultdict a un diccionario normal para facilitar su manejo
    return dict(ventas_mensuales)

def listar_ventas_mensuales(ventas):
    # Imprime las ventas por mes y producto
    print("Ventas mensuales por producto:")
    for mes, productos in ventas.items():
        print(f"\nMes: {mes}")
        for producto, cantidad in productos.items():
            print(f"  {producto}: {cantidad} unidades vendidas")

# Llama a la función ventas_por_mes con los datos cargados y obtiene las ventas mensuales
ventas
````
