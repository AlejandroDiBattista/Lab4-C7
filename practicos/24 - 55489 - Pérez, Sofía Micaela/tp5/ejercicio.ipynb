## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad): pass
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba de la clase Producto

p1 = Producto('0001', 'Coca Cola', 1500, 'gaseosa', 10)

assert p1.codigo == '0001'
assert p1.nombre == 'Coca Cola'
assert p1.precio == 1500

# Calcula el valor total 
assert p1.valorTotal() == 15000 

# Asegura que los valores de precio y cantidad sean validos
p1.precio = -200
assert p1.precio == 1500    # Rechaza el valor negativo

p1.precio = 2000 
assert p1.precio == 2000

p1.cantidad = -5
assert p1.cantidad == 10    # Rechaza el valor negativo

p1.cantidad = 50
assert p1.cantidad == 50
assert p1.valorTotal() == 100000

print("Prueba pasada exitosamente!")
```

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        assert len(codigo) == 4 and codigo.isdigit(), "Código debe ser de 4 dígitos"
        assert 1 <= len(nombre) <= 100, "Nombre debe tener entre 1 y 100 caracteres"
        assert 10 <= precio <= 10000, "Precio debe estar entre 10 y 10,000"
        assert 0 <= len(tipo) <= 20, "Tipo debe tener entre 0 y 20 caracteres"
        assert 0 <= cantidad <= 1000, "Cantidad debe estar entre 0 y 1000"

        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad

    def valor_total(self):
        return self.precio * self.cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self,descripcion, codigos = [], tipos = []): pass 

class OfertaDescuento(Oferta):
    def __init__(self, descuento, *args, **kwargs): pass 
    
class Oferta2x1(Oferta):
    def __init__(self, *args, **kwargs): pass 
```

```python
## NO MODIFICAR ESTE CODIGO ##

p1 = Producto('1234', 'Coca Cola', 1000, 'gaseosa', 10)
p2 = Producto('1235', 'Oreo',      2300, 'galleta', 10)

o10d = OfertaDescuento(10, codigos=['1234'])
assert o10d.calcularDescuento(p1, 10) == 1000 
assert o10d.calcularDescuento(p1, 1) == 100

assert o10d.calcularDescuento(p2, 10) == 0

o2x1 = Oferta2x1(tipos=['galleta'])
assert o2x1.calcularDescuento(p1, 10) == 0

assert o2x1.calcularDescuento(p2, 1) == 0
assert o2x1.calcularDescuento(p2, 2) == 2300
assert o2x1.calcularDescuento(p2, 3) == 2300
assert o2x1.calcularDescuento(p2, 4) == 4600
assert o2x1.calcularDescuento(p2, 5) == 4600

print("Prueba pasada exitosamente!")
```

```python
class Oferta:
    def __init__(self, tipo, descuento, codigo_producto=None, tipo_producto=None):
        self.tipo = tipo
        self.descuento = descuento
        self.codigo_producto = codigo_producto
        self.tipo_producto = tipo_producto

    def es_aplicable(self, producto, cantidad):
        if self.codigo_producto and producto.codigo != self.codigo_producto:
            return False
        if self.tipo_producto and producto.tipo != self.tipo_producto:
            return False
        return True

    def calcular_descuento(self, producto, cantidad):
        if self.tipo == "porcentual":
            return producto.precio * cantidad * self.descuento / 100
        elif self.tipo == "2x1" and cantidad >= 2:
            return producto.precio * (cantidad // 2)
        return 0

    def descripcion(self):
        if self.tipo == "porcentual":
            return f"Descuento del {self.descuento}%"
        elif self.tipo == "2x1":
            return "Oferta 2x1"
        return ""
```

```python
class Catalogo:
    def __init__(self, archivo_csv):
        self.productos = {}
        self.ofertas = []
        self.cargar_catalogo(archivo_csv)

    def cargar_catalogo(self, archivo_csv):
        with open(archivo_csv, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                producto = Producto(row['codigo'], row['nombre'], float(row['precio']), row['tipo'], int(row['cantidad']))
                self.productos[producto.codigo] = producto

    def descontar_existencia(self, codigo, cantidad):
        if codigo in self.productos and self.productos[codigo].cantidad >= cantidad:
            self.productos[codigo].cantidad -= cantidad
        else:
            raise ValueError("Producto no encontrado o cantidad insuficiente")

    def agregar_producto(self, producto):
        self.productos[producto.codigo] = producto

    def buscar_producto(self, codigo):
        return self.productos.get(codigo, None)

    def agregar_oferta(self, oferta):
        self.ofertas.append(oferta)

    def aplicar_oferta(self, codigo, cantidad):
        producto = self.buscar_producto(codigo)
        for oferta in self.ofertas:
            if oferta.es_aplicable(producto, cantidad):
                return oferta.calcular_descuento(producto, cantidad)
        return 0

    def guardar_catalogo(self, archivo_csv):
        with open(archivo_csv, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])
            for producto in self.productos.values():
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])

    def generar_informe(self):
        total_productos = len(self.productos)
        total_unidades = sum([p.cantidad for p in self.productos.values()])
        precio_promedio = sum([p.precio for p in self.productos.values()]) / total_productos if total_productos > 0 else 0
        valor_total = sum([p.valor_total() for p in self.productos.values()])

        tipos = {}
        for p in self.productos.values():
            if p.tipo in tipos:
                tipos[p.tipo]['cantidad'] += p.cantidad
                tipos[p.tipo]['precio_total'] += p.precio * p.cantidad
            else:
                tipos[p.tipo] = {'cantidad': p.cantidad, 'precio_total': p.precio * p.cantidad}

        informe = f"INFORME CATALOGO\nCantidad de productos: {total_productos}\nCantidad de unidades: {total_unidades}\n"
        informe += f"Precio promedio: ${precio_promedio:.2f}\nValor total: ${valor_total:.2f}\nTipos de productos:\n"
        for tipo, data in tipos.items():
            informe += f" - {tipo}: {data['cantidad']}u x ${data['precio_total'] / data['cantidad']:.2f}\n"

        informe += "Ofertas:\n"
        for oferta in self.ofertas:
            informe += f" - {oferta.descripcion()}\n"

        return informe
```

```python
class Cliente:
    def __init__(self, nombre, apellido, cuit):
        assert len(cuit) == 11 and cuit.isdigit(), "CUIT debe tener 11 dígitos"
        self.nombre = nombre
        self.apellido = apellido
        self.cuit = cuit
```

```python
class Factura:
    def __init__(self, catalogo, cliente): pass 
```

```python
class Factura:
    secuencia = 1

    def __init__(self, cliente):
        self.numero = Factura.secuencia
        Factura.secuencia += 1
        self.fecha = datetime.now().strftime("%d/%m/%Y")
        self.cliente = cliente
        self.items = []
        self.subtotal = 0
        self.descuentos = 0

    def agregar_producto(self, producto, cantidad, descuento):
        subtotal_item = producto.precio * cantidad
        self.subtotal += subtotal_item
        self.descuentos += descuento
        self.items.append((producto, cantidad, subtotal_item, descuento))

    def total(self):
        return self.subtotal - self.descuentos

    def generar_texto_factura(self):
        factura = f"Factura: {self.numero}\nFecha : {self.fecha}\nCliente: {self.cliente.nombre} {self.cliente.apellido} ({self.cliente.cuit})\n"
        for item in self.items:
            producto, cantidad, subtotal_item, descuento = item
            factura += f"- {cantidad}u {producto.nombre} x ${producto.precio:.2f} = ${subtotal_item:.2f}\n"
            if descuento > 0:
                factura += f"  Descuento: -${descuento:.2f}\n"
        factura += f"Subtotal: ${self.subtotal:.2f}\nDescuentos: -${self.descuentos:.2f}\n"
        factura += "-----------------------\n"
        factura += f"Total: ${self.total():.2f}\n"
        return factura
```
