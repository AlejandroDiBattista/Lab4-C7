## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = self._comprobar_codigo(codigo)
        self.nombre = self._comprobar_nombre(nombre)
        self._precio = self._comprobar_precio(precio)
        self.tipo = self._comprobar_tipo(tipo)
        self._cantidad = self._comprobar_cantidad(cantidad)

    def _comprobar_codigo(self, codigo):
        if isinstance(codigo, str) and len(codigo) == 4:
            return codigo
        else:
            raise ValueError("El código debe tener 4 caracteres")

    def _comprobar_nombre(self, nombre):
        if 1 <= len(nombre) <= 100:
            return nombre
        else:
            raise ValueError("El nombre debe contener entre 1 y 100 caracteres")

    def _comprobar_precio(self, precio):
        if 10 <= precio <= 10000:
            return precio
        else:
            raise ValueError("El precio debe estar en el rango de 10 a 10,000")

    def _comprobar_tipo(self, tipo):
        if len(tipo) <= 20:
            return tipo
        else:
            raise ValueError("El tipo no debe exceder los 20 caracteres")

    def _comprobar_cantidad(self, cantidad):
        if 0 <= cantidad <= 1000:
            return cantidad
        else:
            raise ValueError("no hay esa cantidad")

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, nuevo_precio):
        try:
            self._precio = self._comprobar_precio(nuevo_precio)
        except ValueError:
            pass

    @property
    def cantidad(self):
        return self._cantidad

    @cantidad.setter
    def cantidad(self, nueva_cantidad):
        try:
            self._cantidad = self._comprobar_cantidad(nueva_cantidad)
        except ValueError:
            pass

    def valorTotal(self):
        return self.precio * self.cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion, codigos=None, tipos=None):
        self.descripcion = self._validar_descripcion(descripcion)
        self.codigos = self._validar_lista(codigos) if codigos is not None else []
        self.tipos = self._validar_lista(tipos) if tipos is not None else []

    def _validar_descripcion(self, descripcion):
        if isinstance(descripcion, str) and 1 <= len(descripcion) <= 100:
            return descripcion
        raise ValueError("La descripción debe tener entre 1 y 100 caracteres.")

    def _validar_lista(self, items):
        return items if items is not None else []

    def es_aplicable(self, producto):
        return producto.codigo in self.codigos or producto.tipo in self.tipos

    def aplicar(self, producto, cantidad):
        raise NotImplementedError("Este método debe ser implementado por las subclases.")


class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion="Descuento", codigos=None, tipos=None):
        super().__init__(descripcion, codigos, tipos)
        self.descuento = self._validar_descuento(descuento)

    def _validar_descuento(self, descuento):
        if 0 < descuento <= 100:
            return descuento
        raise ValueError("El descuento debe ser mayor a 0 y menor o igual a 100.")

    def aplicar(self, producto, cantidad):
        if self.es_aplicable(producto):
            precio_final = producto.precio * (1 - self.descuento / 100)
            return precio_final * cantidad
        return producto.precio * cantidad

    def calcularDescuento(self, producto, cantidad):
        if self.es_aplicable(producto):
            descuento_total = producto.precio * (self.descuento / 100) * cantidad
            return descuento_total
        return 0


class Oferta2x1(Oferta):
    def __init__(self, descripcion="Oferta 2x1", codigos=None, tipos=None):
        super().__init__(descripcion, codigos, tipos)

    def aplicar(self, producto, cantidad):
        if self.es_aplicable(producto):
            cantidad_a_pagar = (cantidad // 2) + (cantidad % 2)
            return producto.precio * cantidad_a_pagar
        return producto.precio * cantidad

    def calcularDescuento(self, producto, cantidad):
        if self.es_aplicable(producto):
            cant_a_pagar = (cantidad // 2) + (cantidad % 2)
            desc_t = producto.precio * (cantidad - cant_a_pagar)
            return desc_t
        return 0
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []
        
    def agregar(self, *productos):
        for p in productos:
            self.productos.append(p)

    def buscar(self, verificarCodigo):
        for p in self.productos:
            if p.codigo == verificarCodigo:
                return p
        return None  

    def registrarOferta(self, of):
        self.ofertas.append(of)

    def buscarOferta(self, producto):
        for of in self.ofertas:
            if of.es_aplicable(producto):
                return of
        return None  

    def calcularDescuento(self, producto, cantidad):
        of = self.buscarOferta(producto)
        if of:
            return of.calcularDescuento(producto, cantidad)
        return 0

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(producto.cantidad for producto in self.productos)

    @property
    def valorTotal(self):
        return sum(producto.precio * producto.cantidad for producto in self.productos)

    def informe(self):
        totalP = self.cantidadProductos
        total_Unid = self.cantidadUnidades
        valor_t = self.valorTotal
        precioP_ = valor_t / total_Unid if total_Unid > 0 else 0.0
        tipos = {}

        for producto in self.productos:
            if producto.tipo not in tipos:
                tipos[producto.tipo] = {'unidades': 0, 'precio_total': 0}
            tipos[producto.tipo]['unidades'] += producto.cantidad
            tipos[producto.tipo]['precio_total'] += producto.precio * producto.cantidad
            
        informe_str = "INFORME CATALOGO\n"
        informe_str += f"Cantidad de productos:   {totalP}\n"
        informe_str += f"Cantidad de unidades:    {total_Unid}\n"
        informe_str += f"Precio Promedio:       $ {precioP_:.2f}\n"
        informe_str += f"Valor total:           $ {valor_t:.2f}\n"
        informe_str += "Tipos de productos: \n"
        for tipo, data in tipos.items():
            unidades = data['unidades']
            precioTipoPromedio = data['precio_total'] / unidades if unidades > 0 else 0.0
            informe_str += f"  - {tipo}              :  {unidades}u x $ {precioTipoPromedio:.2f}\n"

        informe_str += "Ofertas:\n"
        for oferta in self.ofertas:
            informe_str += f"  - {oferta.descripcion}\n"

        return informe_str 

    @classmethod
    def leer(cls, archivo):
        cat = cls()
        with open(archivo, 'r') as f:
            next(f)
            for linea in f:
                codigo, nombre, precio, tipo, cantidad = linea.strip().split(',')
                producto = Producto(codigo, nombre, float(precio), tipo, int(cantidad))  
                cat.agregar(producto)
        return cat

    def vender(self, producto, cantidad):
        if cantidad <= producto.cantidad:
            producto.cantidad -= cantidad
        else:
            raise ValueError("No hay de las cantidades buscadas.")

    def guardar(self, archivo):
        with open(archivo, 'w') as archivo2:
            archivo2.write("codigo,nombre,precio,tipo,cantidad\n")
            for producto in self.productos:
                archivo2.write(f"{producto.codigo},{producto.nombre},{producto.precio},{producto.tipo},{producto.cantidad}\n")
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit):
        self._nombre = self._validar_nombre(nombre)
        self._cuit = self._validar_cuit(cuit)

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, nuevo_nombre):
        if nuevo_nombre == '':
            return
        self._nombre = self._validar_nombre(nuevo_nombre)

    @property
    def cuit(self):
        return self._cuit

    @cuit.setter
    def cuit(self, nuevo_cuit):
        try:
            self._cuit = self._validar_cuit(nuevo_cuit)
        except ValueError:
            pass  # No cambia el CUIT si es inválido

    def _validar_nombre(self, nombre):
        if isinstance(nombre, str) and 1 <= len(nombre) <= 100:
            return nombre
        raise ValueError("El nombre completo debe tener entre 1 y 100 caracteres.")

    def _validar_cuit(self, cuit):
        # Validación del CUIT
        if (len(cuit) == 13 and 
            cuit[2] == '-' and 
            cuit[11] == '-' and 
            cuit[:2].isdigit() and 
            cuit[3:11].isdigit() and 
            cuit[12].isdigit()):
            return cuit  # Devuelve el CUIT válido
        raise ValueError("El CUIT debe tener el formato XX-XXXXXXXX-X y contener 13 dígitos.")

    def __str__(self):
        return f"Cliente(nombre_completo='{self._nombre}', cuit='{self._cuit}')"
```

```python
class Factura:
    _numero = 100  # Número inicial de factura
    _ultima_factura = None  # Para almacenar la última factura creada

    def __init__(self, catalogo, cliente):
        self.catalogo = catalogo
        self.cliente = cliente
        self.numero = self.nuevoNumero()
        self.fecha = "fecha_placeholder"  # Aquí puedes poner la lógica para la fecha
        self.productos = {}  # Diccionario para almacenar los productos y cantidades
        self.descuentos = 0.0
        self.total = 0.0
        Factura._ultima_factura = self  # Actualizamos la última factura creada

    @classmethod
    def nuevoNumero(cls):
        cls._numero += 1
        return cls._numero

    @classmethod
    def ultimaFactura(cls):
        return cls._ultima_factura

    def agregar(self, producto, cantidad):
        if producto.codigo in self.productos:
            self.productos[producto.codigo] += cantidad
        else:
            self.productos[producto.codigo] = cantidad

    @property
    def subtotal(self):
        return sum(producto.precio * cantidad for codigo, cantidad in self.productos.items() for producto in [self.catalogo.buscar(codigo)] if producto)

    def imprimir(self):
        impresion = f"Factura: {self.numero}\n"
        impresion += f"Fecha  : {self.fecha}\n"
        impresion += f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})\n\n"

        for codigo, cantidad in self.productos.items():
            producto = self.catalogo.buscar(codigo)
            if producto:  # Verificamos que el producto existe
                subtotal = producto.precio * cantidad
                impresion += f"- {cantidad} {producto.nombre:<30} x ${producto.precio:.2f} = ${subtotal:.2f}\n"
                
                descuento = self.catalogo.calcularDescuento(producto, cantidad)
                if descuento > 0:
                    tasa_descuento = self.catalogo.calcularDescuento(producto, cantidad)  # Cambia esto a tasa_descuento si está implementado
                    impresion += f"  Descuento {tasa_descuento}% - ${descuento:.2f}\n"
                    
                    oferta = self.catalogo.buscarOferta(producto)
                    if oferta:
                        impresion += f"  {oferta.descripcion:<40} - ${descuento:.2f}\n"

        self.descuentos = sum(self.catalogo.calcularDescuento(self.catalogo.buscar(codigo), cantidad) for codigo, cantidad in self.productos.items())
        self.total = self.subtotal - self.descuentos

        impresion += f"\n                                             Subtotal:   ${self.subtotal:.2f}\n"
        impresion += f"                                             Descuentos: ${self.descuentos:.2f}\n"
        impresion += f"                                             -----------------------\n"
        impresion += f"                                             TOTAL:      ${self.total:.2f}\n"

        return impresion
```
