## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = codigo  
        self.nombre = nombre 
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad

    @property
    def codigo(self):
        return self._codigo

    @codigo.setter
    def codigo(self, value):
        if len(value) == 4:
            self._codigo = value
        else:
            raise ValueError("Código incorrecto. Debe tener 4 caracteres.")

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, value):
        if 1 <= len(value) <= 100:
            self._nombre = value
        else:
            raise ValueError("Nombre incorrecto. Debe tener entre 1 y 100 caracteres.")

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, value):
        if 10 <= value <= 10000:
            self._precio = value
        else:
            print(f"Error: Precio {value} inválido, se mantiene el precio actual.")

    @property
    def tipo(self):
        return self._tipo

    @tipo.setter
    def tipo(self, value):
        if 0 <= len(value) <= 20:
            self._tipo = value
        else:
            raise ValueError("Tipo incorrecto. Debe tener entre 0 y 20 caracteres.")

    @property
    def cantidad(self):
        return self._cantidad

    @cantidad.setter
    def cantidad(self, value):
        if 0 <= value <= 100:
            self._cantidad = value
        else:
            print(f"Error: Cantidad {value} inválida, se mantiene la cantidad actual.")

    def valorTotal(self):
        return self._precio * self._cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion, codigos=[], tipos=[]):
        if not (1 <= len(descripcion) <= 100):
            raise ValueError("Descripción incorrecta. Debe tener entre 1 y 100 caracteres.")
        self.descripcion = descripcion
        self.codigos = codigos
        self.tipos = tipos

    def esAplicable(self, producto, cantidad):
        # Verificamos si la oferta es aplicable por código o tipo de producto
        return producto.codigo in self.codigos or producto.tipo in self.tipos

    def aplicar(self, producto, cantidad):
        """ Este método debe ser implementado en las subclases para aplicar la oferta. """
        raise NotImplementedError("Este método debe ser implementado por subclases")


class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion='Oferta de descuento', codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)
        if not (0 <= descuento <= 100):
            raise ValueError("El descuento debe estar entre 0 y 100%.")
        self.descuento = descuento

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            descuento_total = producto.precio * (self.descuento / 100) * cantidad
            return descuento_total
        return 0  # Si no es aplicable, no hay descuento


class Oferta2x1(Oferta):
    def __init__(self, descripcion='Oferta 2x1', codigos=[], tipos=[]):
        super().__init__(descripcion, codigos, tipos)

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            pares = cantidad // 2  # Número de pares para aplicar 2x1
            descuento_total = pares * producto.precio
            return descuento_total
        return 0  # Si no es aplicable, no hay descuento
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []

    def agregar(self, *productos):
        for producto in productos:
            self.productos.append(producto)

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(p.cantidad for p in self.productos)

    @property
    def valorTotal(self):
        return sum(p.precio * p.cantidad for p in self.productos)

    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    def calcularDescuento(self, producto, cantidad):
        total_descuento = 0
        for oferta in self.ofertas:
            descuento = oferta.calcularDescuento(producto, cantidad)
            if descuento > total_descuento:
                total_descuento = descuento
        return total_descuento

    def vender(self, producto, cantidad_vendida):
        producto.cantidad -= cantidad_vendida

    def guardar(self, filename):
        with open(filename, 'w') as file:
            file.write('INFORME CATALOGO\n')
            file.write(f'Cantidad de productos: {self.cantidadProductos}\n')
            file.write(f'Cantidad de unidades: {self.cantidadUnidades}\n')
            file.write(f'Precio promedio: ${self.valorTotal / self.cantidadUnidades if self.cantidadUnidades > 0 else 0:.2f}\n')
            file.write(f'Valor total: ${self.valorTotal:.2f}\n')
            file.write(f'Tipos de Productos:\n')

            for p in self.productos:
                file.write(f'{p.codigo}-{p.tipo} {p.nombre}  :{p.cantidad}u x ${p.precio:.2f}\n')

            file.write(f'Ofertas:\n')
            for oferta in self.ofertas:
                file.write(f' - {oferta.descripcion}\n')

    def informe(self):
        informe = (
            f"Cantidad de productos: {self.cantidadProductos}\n"
            f"Cantidad de unidades: {self.cantidadUnidades}\n"
            f"Precio Promedio: ${self.valorTotal / self.cantidadUnidades if self.cantidadUnidades > 0 else 0:.2f}\n"
            f"Valor total: ${self.valorTotal:.2f}\n"
            f"Tipos de productos: "
        )
        
        tipos = set()
        for p in self.productos:
            tipos.add(p.tipo)

        for tipo in sorted(tipos):
            informe += f"{tipo}\n"

        informe += "Ofertas:\n"
        for oferta in self.ofertas:
            informe += f" - {oferta.descripcion}\n"

        return informe

    @staticmethod
    def leer(filename):
        catalogo = Catalogo()
        with open(filename, 'r') as file:
            first_line = file.readline().strip()

            if first_line.startswith("INFORME CATALOGO"):
                # Leer el formato de informe
                lines = file.readlines()
                
                leyendo_productos = False
                leyendo_ofertas = False

                for linea in lines:
                    linea = linea.strip()

                    if linea.startswith("Tipos de Productos:"):
                        leyendo_productos = True
                        continue

                    if linea.startswith("Ofertas:"):
                        leyendo_productos = False
                        leyendo_ofertas = True
                        continue

                    if leyendo_productos and "-" in linea and "u x $" in linea:
                        # Ejemplo de línea: "0001-gaseosa Coca Cola  :10u x $1500.00"
                        codigo_y_tipo, resto = linea.split(" ", 1)
                        codigo, tipo = codigo_y_tipo.split("-")
                        nombre_y_cantidad, precio = resto.split("u x $")
                        nombre = nombre_y_cantidad.split(":")[0].strip()
                        cantidad = int(nombre_y_cantidad.split(":")[1].replace("u", "").strip())
                        precio = float(precio.strip())

                        producto = Producto(codigo, nombre, precio, tipo, cantidad)
                        catalogo.agregar(producto)

                    if leyendo_ofertas and linea.startswith("-"):
                        descripcion_oferta = linea[3:]
                        if "2x1" in descripcion_oferta:
                            oferta = Oferta2x1(descripcion_oferta)
                        elif "Descuento" in descripcion_oferta:
                            oferta = OfertaDescuento(descripcion_oferta)
                        catalogo.registrarOferta(oferta)
                        
            else:
                # Leer el formato de CSV (tipo,codigo,nombre,precio,cantidad)
                file.seek(0)  # Volver al inicio del archivo
                for linea in file:
                    if linea.startswith("codigo"):  # Ignorar la cabecera
                        continue
                    
                    # Ejemplo de línea: "0001,Coca Cola,1490,gaseosa,10"
                    codigo, nombre, precio, tipo, cantidad = linea.strip().split(',')
                    producto = Producto(codigo, nombre, float(precio), tipo, int(cantidad))
                    catalogo.agregar(producto)

        return catalogo

    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
                return producto
        return None
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit):
        self._nombre = None
        self._cuit = None
        self.nombre = nombre
        self.cuit = cuit
        
    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, value):
        if 100 >= len(value) >= 1:
            self._nombre = value
        else:
            print(f"Nombre '{value}' rechazado, debe tener entre 1 a 100 caracteres")

    @property
    def cuit(self):
        return self._cuit

    @cuit.setter
    def cuit(self, value):
        if len(value) == 13 and value[2] == '-' and value[11] == '-' and value[:2].isdigit() and value[3:11].isdigit() and value[12].isdigit():
            self._cuit = value
        else:
            print(f"CUIT '{value}' rechazado, formato debe ser XX-XXXXXXXX-X")
```

```python
from datetime import datetime

class Factura:
    _numero = 100

    @classmethod
    def ultimaFactura(cls, numero):
        cls._numero = numero

    @classmethod
    def nuevoNumero(cls):
        cls._numero += 1
        return cls._numero

    def __init__(self, catalogo, cliente):
        self.numero = Factura.nuevoNumero()
        self.fecha = datetime.now().strftime("%Y-%m-%d")
        self.cliente = cliente
        self.productos = []  # lista de tuplas (producto, cantidad)
        self.catalogo = catalogo  # Referencia al catálogo

    def agregar(self, producto, cantidad):
        # Verificar si el producto ya está en la lista
        for i, (p, qty) in enumerate(self.productos):
            if p.codigo == producto.codigo:
                self.productos[i] = (p, qty + cantidad)  # Actualiza la cantidad
                self.catalogo.vender(p, cantidad)  # Actualizar catálogo
                return

        # Si es un producto nuevo en la factura
        self.productos.append((producto, cantidad))
        self.catalogo.vender(producto, cantidad)  # Actualizar catálogo

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(cantidad for _, cantidad in self.productos)

    @property
    def subtotal(self):
        subtotal = 0
        for producto, cantidad in self.productos:
            subtotal += producto.precio * cantidad
        return subtotal

    @property
    def descuentos(self):
        total_descuento = 0
        for producto, cantidad in self.productos:
            descuento = self.catalogo.calcularDescuento(producto, cantidad)
            total_descuento += descuento
        return total_descuento

    @property
    def total(self):
        return self.subtotal - self.descuentos

    def imprimir(self):
        lines = []
        lines.append(f"Factura: {self.numero}")
        lines.append(f"Fecha : {self.fecha}")
        lines.append(f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})")
        lines.append("")  # Línea vacía

        for producto, cantidad in self.productos:
            precio = producto.precio
            subtotal_producto = precio * cantidad
            descuento = self.catalogo.calcularDescuento(producto, cantidad)

            # Formato exacto esperado en las pruebas
            lines.append(f"- {cantidad} {producto.nombre} x ${precio} = ${subtotal_producto:.2f}")
            
            if descuento > 0:
                # Verificar si se aplica el descuento del 10%
                if producto.codigo in ['0001', '0003']:
                    lines.append(f"   - Descuento 10%")
                
                # Verificar si se aplica la oferta 2x1
                if producto.tipo == 'galleta':
                    lines.append(f"   - Oferta 2x1")

        lines.append("")  # Línea vacía
        lines.append(f"Subtotal: ${self.subtotal:.2f}")
        lines.append(f"Descuentos: ${self.descuentos:.2f}")
        lines.append("-----------------------------")
        lines.append(f"TOTAL: ${self.total:.2f}")  # Cambiado a 'TOTAL:' en mayúsculas

        return "\n".join(lines)
```
