## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = self.validar_codigo(codigo)
        self.nombre = self.validar_nombre(nombre)
        self.precio = self.validar_precio(precio)
        self.tipo = self.validar_tipo(tipo)
        self.cantidad = self.validar_cantidad(cantidad)
        self.valorTotal = self.calcular_valor_total()


    def validar_codigo(self, codigo):
        if idinstance(codigo, int) and 1000 <= codigo <= 9999:
            return codigo
        else:
            raise ValueError("El código debe ser un número entero de 4 dígitos.")

    def validar_nombre(self, nombre):
        if idinstance(nombre, str) and 1 <= len(nombre) <= 100:
            return nombre
        else:
            raise ValueError("El nombre debe tener entre 1 y 100 caracteres.")

    def validar_precio(self, precio):
        if idinstance(precio, (int, float)) and 10 <= precio <= 10000:
            return precio
        else:
            raise ValueError("El precio debe estar entre 10 y 10,000.")
    def validar_tipo(self, tipo):
        if idinstance(tipo, str) and 0 <= len(tipo) <= 20:
            return tipo
        else:
            raise ValueError("El tipo debe tener entre 0 y 20 caracteres.")

    
    def validar_cantidad(self, cantidad):
        if isinstance(cantidad, int) and 0 <= cantidad <= 100:
            return cantidad
        else:
            raise ValueError("La cantidad debe estar entre 0 y 100.")

    
    def calcular_valor_total(self):
        return self.cantidad * self.precio

    def __str__(self):
        return f"Producto({self.codigo}, {self.nombre}, {self.precio}, {self.tipo}, {self.cantidad}, Valor Total: {self.valorTotal})"



try:
    producto = Producto(1234, "Laptop", 1500, "Electrónica", 10)
    print(producto)
except ValueError as e:
    print(e)
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion, codigos, tipos, descuento):
        self.descripcion = self.validar_descripcion(descripcion)
        self.codigos = self.validar_codigos(codigos)
        self.tipos = self.validar_tipos(tipos)
        self.descuento = self.validar_descuento(descuento)  

    
    def validar_descripcion(self, descripcion):
        if idinstance(descripcion, str) and 1 <= len(descripcion) <= 100:
            return descripcion
        else:
            raise ValueError("La descripción debe tener entre 1 y 100 caracteres.")

    
    def validar_codigos(self, codigos):
        if all(idinstance(codigo, int) and 1000 <= codigo <= 9999 for codigo in codigos):
            return codigos
        else:
            raise ValueError("Todos los códigos deben ser números enteros de 4 dígitos.")

    def validar_tipos(self, tipos):
        if all(idinstance(tipo, str) and 0 <= len(tipo) <= 20 for tipo in tipos):
            return tipos
        else:
            raise ValueError("Todos los tipos deben ser cadenas de texto con máximo 20 caracteres.")

  
    def validar_descuento(self, descuento):
        if idinstance(descuento, (int, float)) and 0 <= descuento <= 100:
            return descuento
        else:
            raise ValueError("El descuento debe estar entre 0 y 100%.")

    
    def esAplicable(self, producto, cantidad):

        if producto.codigo in self.codigos or producto.tipo in self.tipos:
            return True
        return False

    
    def aplicar(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            precio_final = producto.precio * (1 - self.descuento / 100)
            return precio_final * cantidad  
        else:
            return producto.precio * cantidad  


class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad


producto1 = Producto(1234, "Laptop", 1500, "Electrónica", 1)
producto2 = Producto(5678, "Teléfono", 800, "Electrónica", 2)
producto3 = Producto(9012, "Cámara", 600, "Fotografía", 1)


oferta = Oferta("Descuento en Electrónica", [1234, 5678], ["Electrónica"], 10)


for producto in [producto1, producto2, producto3]:
    if oferta.esAplicable(producto, producto.cantidad):
        print(f"Oferta aplicable a {producto.nombre}. Precio final: {oferta.aplicar(producto, producto.cantidad)}")
    else:
        print(f"No hay oferta disponible para {producto.nombre}. Precio final: {producto.precio * producto.cantidad}")
```
