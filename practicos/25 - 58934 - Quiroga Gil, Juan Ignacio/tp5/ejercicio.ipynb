## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self._codigo = self._validar_codigo(codigo)
        self._nombre = self._validar_nombre(nombre)
        self._precio = self._validar_precio(precio)
        self._tipo = self._validar_tipo(tipo)
        self._cantidad = self._validar_cantidad(cantidad)

    @property
    def codigo(self):
        return self._codigo

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, value):
        nuevo_nombre = self._validar_nombre(value)
        if nuevo_nombre:
            self._nombre = nuevo_nombre

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, value):
        nuevo_precio = self._validar_precio(value)
        if nuevo_precio:
            self._precio = nuevo_precio

    @property
    def tipo(self):
        return self._tipo

    @tipo.setter
    def tipo(self, value):
        nuevo_tipo = self._validar_tipo(value)
        if nuevo_tipo:
            self._tipo = nuevo_tipo

    @property
    def cantidad(self):
        return self._cantidad

    @cantidad.setter
    def cantidad(self, value):
        nueva_cantidad = self._validar_cantidad(value)
        if nueva_cantidad is not None:
            self._cantidad = nueva_cantidad

    def valorTotal(self):
        return self._precio * self._cantidad

    def _validar_codigo(self, codigo):
        if not isinstance(codigo, str) or len(codigo) != 4 or not codigo.isdigit():
            raise ValueError("El código debe ser una cadena de 4 dígitos")
        return codigo

    def _validar_nombre(self, nombre):
        if isinstance(nombre, str) and 1 <= len(nombre) <= 100:
            return nombre
        return None

    def _validar_precio(self, precio):
        if isinstance(precio, (int, float)) and 10 <= precio <= 10000:
            return precio
        return None

    def _validar_tipo(self, tipo):
        if isinstance(tipo, str) and len(tipo) <= 20:
            return tipo
        return None

    def _validar_cantidad(self, cantidad):
        if isinstance(cantidad, int) and 0 <= cantidad <= 1000:
            return cantidad
        return None
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion, codigos=None, tipos=None):
        self.descripcion = self._validar_descripcion(descripcion)
        self.codigos = codigos or []
        self.tipos = tipos or []

    def esAplicable(self, producto, cantidad):
        return (producto.codigo in self.codigos) or (producto.tipo in self.tipos)

    def aplicar(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return self.calcularDescuento(producto, cantidad)
        return 0

    def calcularDescuento(self, producto, cantidad):
        raise NotImplementedError("Debe ser implementado por las subclases")

    def _validar_descripcion(self, descripcion):
        if not isinstance(descripcion, str) or len(descripcion) < 1 or len(descripcion) > 100:
            raise ValueError("La descripción debe ser una cadena de 1 a 100 caracteres")
        return descripcion

class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion=None, codigos=None, tipos=None):
        if descripcion is None:
            descripcion = f"Descuento del {descuento}%"
        super().__init__(descripcion, codigos=codigos, tipos=tipos)
        self.descuento = descuento

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return (producto.precio * cantidad * self.descuento) / 100
        return 0

class Oferta2x1(Oferta):
    def __init__(self, descripcion="Oferta 2x1", codigos=None, tipos=None):
        super().__init__(descripcion, codigos=codigos, tipos=tipos)

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto, cantidad):
            return producto.precio * (cantidad // 2)
        return 0

class Producto:
    def __init__(self, codigo, nombre, precio, tipo, stock):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.stock = stock

# Test cases
p1 = Producto('1234', 'Coca Cola', 1000, 'gaseosa', 10)
p2 = Producto('1235', 'Oreo', 2300, 'galleta', 10)

o10d = OfertaDescuento(10, codigos=['1234'])
assert o10d.calcularDescuento(p1, 10) == 1000
assert o10d.calcularDescuento(p1, 1) == 100
assert o10d.calcularDescuento(p2, 10) == 0

o2x1 = Oferta2x1(tipos=['galleta'])
assert o2x1.calcularDescuento(p1, 10) == 0
assert o2x1.calcularDescuento(p2, 10) == 11500
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv


class Catalogo:
    def __init__(self):
        self.productos = []  
        self.ofertas = []    

    @classmethod
    def leer(cls, archivo):
        catalogo = cls()
        with open(archivo, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                producto = Producto(
                    row['codigo'],
                    row['nombre'],
                    float(row['precio']),
                    row['tipo'],
                    int(row['cantidad'])
                )
                catalogo.productos.append(producto)
        return catalogo
    
    def guardar(self, archivo):
        with open(archivo, 'w', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])
            for producto in self.productos:
                writer.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])

    def agregar(self, producto):
        self.productos.append(producto)

    def buscar(self, codigo):
        for producto in self.productos:
            if producto.codigo == codigo:
             return producto
        return None

    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    def calcularDescuento(self, producto, cantidad):
        descuento_maximo = 0
        for oferta in self.ofertas:
            descuento_actual = oferta.aplicar(producto, cantidad)
            if descuento_actual > descuento_maximo:
                descuento_maximo = descuento_actual
        return descuento_maximo

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(producto.cantidad for producto in self.productos)

    @property
    def valorTotal(self):
        return sum(producto.precio * producto.cantidad for producto in self.productos)

    def vender(self, producto, cantidad):
        prod = self.buscar(producto.codigo)
        if prod and prod.cantidad >= cantidad:
            prod.cantidad -= cantidad

    def informe(self):
        total_unidades = self.cantidadUnidades
        total_valor = self.valorTotal
        total_productos = self.cantidadProductos

        if total_unidades > 0:
            precio_promedio = total_valor / total_unidades
        else:
            precio_promedio = 0

        tipos_info = ""
        tipos_productos = {}
        for producto in self.productos:
                if producto.tipo not in tipos_productos:
                    tipos_productos[producto.tipo] = {'unidades': 0, 'precio_total': 0}
                tipos_productos[producto.tipo]['unidades'] += producto.cantidad
                tipos_productos[producto.tipo]['precio_total'] += producto.precio * producto.cantidad

        for tipo, datos in tipos_productos.items():
            precio_promedio_tipo = datos['precio_total'] / datos['unidades'] if datos['unidades'] > 0 else 0
            tipos_info += f"  - {tipo} : {datos['unidades']}u x $ {precio_promedio_tipo:.2f}\n"

        ofertas_info = ""
        for oferta in self.ofertas:
            if isinstance(oferta, Oferta2x1):
                ofertas_info += "  - Oferta 2x1\n"
            elif isinstance(oferta, OfertaDescuento):
                ofertas_info += f"  - Oferta Descuento {oferta.porcentaje}%\n"
            else:
                ofertas_info += f"  - {oferta.__class__.__name__}\n"

        informe = (
            "INFORME CATALOGO\n"
            f"Cantidad de productos: {total_productos}\n"
            f"Cantidad de unidades: {total_unidades}\n"
            f"Precio Promedio: $ {precio_promedio:.2f}\n"
            f"Valor total: $ {total_valor:.2f}\n"
            f"Tipos de productos: \n{tipos_info}"
            f"Ofertas:\n{ofertas_info}"
        )

        return informe
    
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad

class Oferta2x1:
    def __init__(self, tipos):
        self.tipos = tipos

    def aplicar(self, producto, cantidad):

        if producto.tipo in self.tipos:

            unidades_gratis = cantidad // 2
            return unidades_gratis * producto.precio
        return 0

class OfertaDescuento:
    def __init__(self, porcentaje, codigos=[]):
        self.porcentaje = porcentaje
        self.codigos = codigos

    def aplicar(self, producto, cantidad):
        if producto.codigo in self.codigos:
            return cantidad * producto.precio * (self.porcentaje / 100)
        return 0
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
import re

class Cliente: 
    def __init__(self, nombre, cuit):
        self._nombre = None
        self._cuit = None
        self.nombre = nombre 
        self.cuit = cuit      

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, value):
        if not value or len(value) < 1 or len(value) > 100:
            return 
        self._nombre = value

    @property
    def cuit(self):
        return self._cuit

    @cuit.setter
    def cuit(self, value):
     
        if not re.match(r'^\d{2}-\d{8}-\d$', value):
            return  
        self._cuit = value
```

```python
import datetime
from typing import List, Dict

class Producto:
    def __init__(self, codigo: str, nombre: str, precio: float, tipo: str, stock: int):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = float(precio)
        self.tipo = tipo
        self.stock = int(stock)

class Catalogo:
    def __init__(self):
        self.productos: Dict[str, Producto] = {}
        self.ofertas: List[Oferta] = []

    def agregar(self, *productos: Producto):
        for producto in productos:
            self.productos[producto.codigo] = producto

    def buscar(self, codigo: str) -> Producto:
        return self.productos.get(codigo)

    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    @property
    def cantidadProductos(self):
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        return sum(p.stock for p in self.productos.values())

    @property
    def valorTotal(self):
        return sum(p.precio * p.stock for p in self.productos.values())

    @classmethod
    def leer(cls, archivo: str):
        catalogo = cls()
        with open(archivo, 'r') as f:
            next(f)  # Skip header
            for line in f:
                codigo, nombre, precio, tipo, cantidad = line.strip().split(',')
                producto = Producto(codigo, nombre, float(precio), tipo, int(cantidad))
                catalogo.agregar(producto)
        return catalogo

class Oferta:
    def aplica(self, producto: Producto) -> bool:
        raise NotImplementedError

    def descuento(self, producto: Producto, cantidad: int) -> float:
        raise NotImplementedError

    @property
    def descripcion(self) -> str:
        raise NotImplementedError

class Oferta2x1(Oferta):
    def __init__(self, tipos: List[str] = None, codigos: List[str] = None):
        self.tipos = tipos or []
        self.codigos = codigos or []

    def aplica(self, producto: Producto) -> bool:
        return producto.tipo in self.tipos or producto.codigo in self.codigos

    def descuento(self, producto: Producto, cantidad: int) -> float:
        return (cantidad // 2) * producto.precio

    @property
    def descripcion(self) -> str:
        return "Oferta 2x1"

class OfertaDescuento(Oferta):
    def __init__(self, porcentaje: float, tipos: List[str] = None, codigos: List[str] = None):
        self.porcentaje = porcentaje
        self.tipos = tipos or []
        self.codigos = codigos or []

    def aplica(self, producto: Producto) -> bool:
        return producto.tipo in self.tipos or producto.codigo in self.codigos

    def descuento(self, producto: Producto, cantidad: int) -> float:
        return round((self.porcentaje / 100) * producto.precio * cantidad, 2)

    @property
    def descripcion(self) -> str:
        return f"Descuento {self.porcentaje}%"

class Cliente:
    def __init__(self, nombre: str, cuit: str):
        self.nombre = nombre
        self.cuit = cuit

class Factura:
    _ultimo_numero = 0

    @classmethod
    def ultimaFactura(cls, numero):
        cls._ultimo_numero = numero

    @classmethod
    def nuevoNumero(cls):
        cls._ultimo_numero += 1
        return cls._ultimo_numero

    def __init__(self, catalogo, cliente):
        self.numero = Factura.nuevoNumero()
        self.fecha = datetime.date.today()
        self.cliente = cliente
        self.catalogo = catalogo
        self.items = {}
        self.cantidadProductos = 0
        self.cantidadUnidades = 0

    def imprimir(self):
        output = []
        output.append(f"Factura: {self.numero}")
        output.append(f"Fecha  : {self.fecha}")
        output.append(f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})")
        output.append("")

        for item in self.items.values():
            producto = item['producto']
            cantidad = item['cantidad']
            subtotal = producto.precio * cantidad
            output.append(f"- {cantidad} {producto.nombre:<20} x ${producto.precio:.2f} = ${subtotal:.2f}")
            
            for oferta in self.catalogo.ofertas:
                if oferta.aplica(producto):
                    descuento = oferta.descuento(producto, cantidad)
                    if descuento > 0:
                        output.append(f"      {oferta.descripcion:<35} - ${descuento:.2f}")

        output.append("")
        output.append(f"{'Subtotal:':<45} ${self.subtotal:.2f}")
        output.append(f"{'Descuentos:':<45} ${self.descuentos:.2f}")
        output.append("                                             -----------------------")
        output.append(f"{'TOTAL:':<45} ${self.total:.2f}")

        return "\n".join(output)

    def agregar(self, producto, cantidad):
        if producto.codigo in self.items:
            self.items[producto.codigo]['cantidad'] += cantidad
        else:
            self.items[producto.codigo] = {'producto': producto, 'cantidad': cantidad}
            self.cantidadProductos += 1
        self.cantidadUnidades += cantidad
        producto.stock -= cantidad

    @property
    def subtotal(self):
        return sum(item['producto'].precio * item['cantidad'] for item in self.items.values())

    @property
    def descuentos(self):
        total_descuento = 0
        for item in self.items.values():
            producto = item['producto']
            cantidad = item['cantidad']
            descuento_maximo = 0
            for oferta in self.catalogo.ofertas:
                if oferta.aplica(producto):
                    descuento_oferta = oferta.descuento(producto, cantidad)
                    descuento_maximo = max(descuento_maximo, descuento_oferta)
            total_descuento += descuento_maximo
        return total_descuento

    @property
    def total(self):
        return self.subtotal - self.descuentos
    
    
```
