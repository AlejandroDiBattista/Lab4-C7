## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def _init_(self, codigo, nombre, precio, tipo, cantidad):
        self.codigo = self.validarCodigo(codigo)
        self.nombre = self.validarNombre(nombre)
        self._precio = self.validarPrecio(precio)
        self.tipo = self.validarTipo(tipo)
        self._cantidad = self.validarCantidad(cantidad)

    def validarCodigo(self, codigo):
        if len(codigo) == 4:
            return codigo
        raise ValueError("Error: Longitud incorrecta de números")

    def validarNombre(self, nombre):
        if len(nombre) >= 1 and len(nombre) <= 100:
            return nombre
        raise ValueError("Error: Longitud incorrecta de caracteres")

    def validarPrecio(self, precio):
        if precio >= 10 and precio <= 10000:
            return precio
        raise ValueError("Error: El precio debe estar entre 10 y 10,000.")

    def validarTipo(self, tipo):
        if len(tipo) <= 20:
            return tipo
        raise ValueError("Error: Longitud incorrecta de caracteres para la categoría")

    def validarCantidad(self, cantidad):
        if cantidad >= 0 and cantidad <= 1000:
            return cantidad
        raise ValueError("Error: Valor incorrecto para el cantidad")

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, valor):
        try:
            nuevo_precio = self.comprobarprecio(valor)
            self._precio = nuevo_precio
        except ValueError:
            pass

    @property
    def cantidad(self):
        return self._cantidad

    @cantidad.setter
    def cantidad(self, valor):
        try:
            nuevo_cantidad = self.comprobarcantidad(valor)
            self._cantidad = nuevo_cantidad
        except ValueError:
            pass

    def calcularValorTotal(self):
        return self.precio * self.cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def _init_(self, descripcion, codigos=None, tipos=None):
        self.descripcion = self.verificardescripcion(descripcion)
        self.codigos = codigos or []
        self.tipos = tipos or []

    def verificardescripcion(self, descripcion):
        if 1 <= len(descripcion) <= 100:
            return descripcion
        raise ValueError("Error: La longitud del descripcion debe estar entre 1 y 100 caracteres.")

    def aplicaAProducto(self, item):
        return item.codigo in self.codigos or item.tipo in self.tipos

    def aplicarPromocion(self, item, cantidad):
        raise NotImplementedError("Error: Este método no ha sido implementado.")


class OfertaDescuento(Oferta):
    def _init_(self, descuento, descripcion="Descuento de 10%", codigos=None, tipos=None):
        super()._init_(descripcion, codigos, tipos)
        self.descuento = self.verificardescuento(descuento)

    def verificardescuento(self, descuento):
        if 0 < descuento <= 100:
            return descuento
        raise ValueError("Error: El descuento debe ser entre 1 y 100.")

    def aplicarPromocion(self, item, cantidad):
        if self.aplicaAProducto(item):
            precioReducido = item.precio * (1 - self.descuento / 100)
            return precioReducido * cantidad
        return item.precio * cantidad

    def calcularTotaldescuento(self, item, cantidad):
        if self.aplicaAProducto(item):
            totaldescuento = item.precio * self.descuento / 100 * cantidad
            return totaldescuento
        return 0


class Oferta2x1(Oferta):
    def _init_(self, descripcion="Oferta 2x1", codigos=None, tipos=None):
        super()._init_(descripcion, codigos, tipos)

    def aplicarPromocion(self, item, cantidad):
        if self.aplicaAProducto(item):
            pagarCantidad = (cantidad // 2) + (cantidad % 2)
            return item.precio * pagarCantidad
        return item.precio * cantidad

    def calcularTotaldescuento(self, item, cantidad):
        if self.aplicaAProducto(item):
            pagarCantidad = (cantidad // 2) + (cantidad % 2)
            totaldescuento = item.precio * (cantidad - pagarCantidad)
            return totaldescuento
        return 0
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = []
        self.ofertas = []

    def añadir(self, *productos):  # Permite agregar múltiples artículos
        for articulo in productos:
            self.productos.append(articulo)

    def localizar(self, id_articulo):
        for articulo in self.productos:
            if articulo.codigo == id_articulo:
                return articulo
        return None

    def incluirPromocion(self, promocion):
        self.ofertas.append(promocion)

    def localizarPromocion(self, articulo):
        for promocion in self.ofertas:
            if promocion.esValida(articulo):
                return promocion
        return None

    def calcularRebaja(self, articulo, cantidad):
        promocion = self.localizarPromocion(articulo)
        if promocion:
            return promocion.aplicarDescuento(articulo, cantidad)
        return 0

    @property
    def totalproductos(self):
        return len(self.productos)

    @property
    def totalUnidades(self):
        return sum(articulo.cantidad for articulo in self.productos)

    @property
    def montoTotal(self):
        return sum(articulo.precio * articulo.cantidad for articulo in self.productos)

    def reporte(self):
        totalproductos = self.totalproductos
        totalUnidades = self.totalUnidades
        montoTotal = self.montoTotal
        precioPromedio = montoTotal / totalUnidades if totalUnidades > 0 else 0.0
        categorias = {}

        for articulo in self.productos:
            if articulo.tipo not in categorias:
                categorias[articulo.tipo] = {'unidades': 0, 'precio_total': 0}
            categorias[articulo.tipo]['unidades'] += articulo.cantidad
            categorias[articulo.tipo]['precio_total'] += articulo.precio * articulo.cantidad

        reporte_str = "REPORTE INVENTARIO\n"
        reporte_str += f"Total de artículos:   {totalproductos}\n"
        reporte_str += f"Total de unidades:    {totalUnidades}\n"
        reporte_str += f"Precio Promedio:       $ {precioPromedio:.2f}\n"
        reporte_str += f"Valor total:           $ {montoTotal:.2f}\n"
        reporte_str += "Categorías de artículos: \n"
        for tipo, datos in categorias.items():
            unidades = datos['unidades']
            promedioPrecioCategoria = datos['precio_total'] / unidades if unidades > 0 else 0.0
            reporte_str += f"  - {tipo}              :  {unidades}u x $ {promedioPrecioCategoria:.2f}\n"

        reporte_str += "ofertas:\n"
        for promocion in self.ofertas:
            reporte_str += f"  - {promocion.descripcion}\n"

        return reporte_str

    @classmethod
    def cargar(cls, archivo):
        inventario = cls()
        with open(archivo, 'r') as f:
            next(f)
            for linea in f:
                codigo, nombre, precio, tipo, cantidad = linea.strip().split(',')
                articulo = Producto(codigo, nombre, float(precio), tipo, int(cantidad))
                inventario.añadir(articulo)
        return inventario

    def despachar(self, articulo, cantidad):
        if cantidad <= articulo.cantidad:
            articulo.cantidad -= cantidad
        else:
            raise ValueError("Cantidad insuficiente disponible o el artículo no existe.")

    def guardarArchivo(self, archivo):
        with open(archivo, 'w') as f:
            f.write("codigo,nombre,precio,tipo,cantidad\n")
            for articulo in self.productos:
                f.write(f"{articulo.codigo},{articulo.nombre},{articulo.precio},{articulo.tipo},{articulo.cantidad}\n")
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit): 
        self._nombre = None
        self._cuit = None
        self.nombre = nombre
        self.cuit = cuit

    @property
    def nombre_fiscal(self):
        return self._nombre_fiscal

    @nombre_fiscal.setter
    def nombre_fiscal(self, valor):
        if valor and isinstance(valor, str) and 1 <= len(valor) <= 100:
            self._nombre_fiscal = valor

    @property
    def identificacion_fiscal(self):
        return self._identificacion_fiscal

    @identificacion_fiscal.setter
    def identificacion_fiscal(self, valor):
        if self.validar_identificacion_fiscal(valor):
            self._identificacion_fiscal = valor

    def validar_identificacion_fiscal(self, identificacion_fiscal):
        if isinstance(identificacion_fiscal, str) and len(identificacion_fiscal) == 13:
            if identificacion_fiscal[2] == '-' and identificacion_fiscal[11] == '-':
                return all(part.isdigit() for part in (identificacion_fiscal[:2], identificacion_fiscal[3:11], identificacion_fiscal[12:]))
        return False
```

```python
class Factura:
    def __init__(self, catalogo, cliente): 
     @classmethod
     def generarNuevoNumero(cls):
        cls._numeroActual += 1
        return cls._numeroActual

    @classmethod
    def actualizarUltimoNumero(cls, numero):
        cls._numeroActual = numero

    def __init__(self, inventario, cliente):
        self.numero = self.generarNuevoNumero()
        self.fecha = self.obtenerFecha()
        self.cliente = cliente
        self.inventario = inventario
        self.articulos = {}

    def obtenerFecha(self):
        from datetime import datetime
        return datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    def agregarArticulo(self, producto, cantidad):
        if producto.codigo in self.articulos:
            self.articulos[producto.codigo] += cantidad
        else:
            self.articulos[producto.codigo] = cantidad
        
        self.inventario.venderProducto(producto, cantidad)

    @property
    def totalArticulos(self):
        return len(self.articulos)

    @property
    def totalUnidades(self):
        return sum(self.articulos.values())

    @property
    def calcularSubtotal(self):
        total = 0
        for codigo, cantidad in self.articulos.items():
            producto = self.inventario.encontrarProducto(codigo)
            total += producto.precio * cantidad
        return total

    @property
    def calcularDescuentos(self):
        totalDescuento = 0
        for codigo, cantidad in self.articulos.items():
            producto = self.inventario.encontrarProducto(codigo)
            descuento = self.inventario.aplicarDescuento(producto, cantidad)
            totalDescuento += descuento
        return totalDescuento

    @property
    def calcularTotal(self):
        return self.calcularSubtotal - self.calcularDescuentos

    def mostrarRecibo(self):
        recibo_str = f"Recibo: {self.numero}\n"
        recibo_str += f"Fecha  : {self.fecha}\n"
        recibo_str += f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})\n\n"

        for codigo, cantidad in self.articulos.items():
            producto = self.inventario.encontrarProducto(codigo)
            subtotal = producto.precio * cantidad
            recibo_str += f"- {cantidad} {producto.nombre:<30} x ${producto.precio:.2f} = ${subtotal:.2f}\n"
            
            descuento = self.inventario.aplicarDescuento(producto, cantidad)
            if descuento > 0:
                oferta = self.inventario.verificarOferta(producto)
                if oferta:
                    recibo_str += f"  {oferta.descripcion:<40} - ${descuento:.2f}\n"

        recibo_str += f"\n                                             Subtotal:   ${self.calcularSubtotal:.2f}\n"
        recibo_str += f"                                             Descuentos: ${self.calcularDescuentos:.2f}\n"
        recibo_str += f"                                             -----------------------\n"
        recibo_str += f"                                             TOTAL:      ${self.calcularTotal:.2f}\n"

        return recibo_str
```
