## Ejercicio 1 

Escribe una función en Python que procese una lista de cadenas de texto y realice las siguientes acciones:

1. Tomar una lista de frases y convertir cada frase en una lista de palabras.
1. Luego, debe crear una nueva lista que contenga la cantidad de palabras en cada frase.
1. Tambien debe crear un lista que contenga la cantidad de caracteres en cada frase.
1. Finalmente, debe imprimir cada frase original junto con la cantidad de palabras que contiene.
Ejemplo de uso:
```python
    frases = [
        "Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"
    ]

    procesar_frases(frases)
```

ejemplo de salida:
```bash
    La frase: "Python es un..."
    tiene 6 palabras y 38 caracteres
```


````python
# Implementar
# ejercicio1.py

def procesar_frases(frases):
    """
    Toma una lista de frases, convierte cada frase en una lista de palabras,
    cuenta la cantidad de palabras y caracteres, e imprime el resultado.
    """
    for frase in frases:
        palabras = frase.split()
        cantidad_palabras = len(palabras)
        cantidad_caracteres = len(frase)
        
        # Imprimir resultados para cada frase
        print(f'La frase: "{frase}"')
        print(f'    tiene {cantidad_palabras} palabras y {cantidad_caracteres} caracteres')

# Ejemplo de uso
if __name__ == "__main__":
    frases = [
        "Python es un lenguaje de programación",
        "Me gusta resolver problemas con código",
        "Las listas y los bucles son muy útiles"
    ]
    
    procesar_frases(frases)

````
