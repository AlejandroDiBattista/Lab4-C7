# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python

def calcular_valor(numero, palo):
    
    if numero == 1 and palo == "espada":
        return 14  
    elif numero == 1 and palo == "basto":
        return 13  
    elif numero == 7 and palo == "espada":
        return 12 
    elif numero == 7 and palo == "oro":
        return 11  
    
    
    elif numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 7:
        return 8
    
    # Cartas con jerarquía menor: 
    elif numero == 6:
        return 7
    elif numero == 5:
        return 6
    elif numero == 4:
        return 5
    
    #  las cartas 12, 11, 10 y 1 de copas/oros
    elif numero == 12:
        return 4
    elif numero == 11:
        return 3
    elif numero == 10:
        return 2
    else:
        return 1  

# Clase Carta 
class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"Carta: {self.numero} de {self.palo} con valor {self.valor}"

# Ejemplo 
c = Carta(1, "oro")
print(c)  # Carta: 1 de oro con valor 1

c2 = Carta(1, "espada")
print(c2)  # Carta: 1 de espada con valor 14
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random


class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"Carta: {self.numero} de {self.palo} con valor {self.valor}"


class Mazo:
    def __init__(self):
        self.cartas = []
        palos = ["oro", "copa", "espada", "basto"]
        numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]  
        
        for palo in palos:
            for numero in numeros:
                self.cartas.append(Carta(numero, palo))
    
    
    def mezclar(self):
        random.shuffle(self.cartas)
    
    
    def repartir(self, cantidad):
        if cantidad > len(self.cartas):
            raise ValueError("No hay suficientes cartas para repartir.")
        
        mano = self.cartas[:cantidad]
        
        self.cartas = self.cartas[cantidad:]
        return mano
    
    def __str__(self):
        return f"Mazo con {len(self.cartas)} cartas restantes"

# Ejemplo 
m = Mazo()
print(f"Antes de mezclar: {m}")  
m.mezclar()
print(f"Después de mezclar: {m}")

# Repartir 3 cartas
mano = m.repartir(3)
print("Cartas repartidas:")
for carta in mano:
    print(carta)

print(f"Después de repartir: {m}")  # mazo con 37 cartas restantes
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
import random

# Clase Jugador 
class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.mano = []  # Cartas 
        self.puntos = 0  # Puntos
    
    #  lista de cartas
    def recibir_cartas(self, cartas):
        self.mano = cartas
    
    # carta aleatoria 
    def jugar_carta(self):
        if not self.mano:
            raise ValueError("No hay cartas para jugar.")
        carta_a_jugar = random.choice(self.mano)
        self.mano.remove(carta_a_jugar)  # Elimina la carta 
        return carta_a_jugar
    
    # Acumula
    def acumular_puntos(self):
        self.puntos += 1
    
    def __str__(self):
        return f"Jugador {self.nombre} con {self.puntos} puntos."

# Ejemplo 
j = Jugador("Juan")
m = Mazo()
m.mezclar()

# Repartimos 
j.recibir_cartas(m.repartir(3))

# Jugador juega una carta
print(f"Cartas en mano: {[str(carta) for carta in j.mano]}")
print("Carta jugada:", j.jugar_carta())
print(f"Cartas restantes en la mano: {[str(carta) for carta in j.mano]}")
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
import random

class Partida:
    def __init__(self, jugador1, jugador2):
        self.jugador1 = jugador1  # Primer jugador
        self.jugador2 = jugador2  # Segundo jugador
        self.mazo = Mazo()  # Crear 
        self.puntaje_maximo = 15  # Puntaje 

    
    def jugar_ronda(self):
        carta_j1 = self.jugador1.jugar_carta()
        carta_j2 = self.jugador2.jugar_carta()
        
        print(f"{self.jugador1.nombre} juega {carta_j1.numero} de {carta_j1.palo}")
        print(f"{self.jugador2.nombre} juega {carta_j2.numero} de {carta_j2.palo}")

        #  ganador 
        if carta_j1.valor > carta_j2.valor:
            print(f"Gana {self.jugador1.nombre} la ronda.")
            self.jugador1.acumular_puntos()
        elif carta_j1.valor < carta_j2.valor:
            print(f"Gana {self.jugador2.nombre} la ronda.")
            self.jugador2.acumular_puntos()
        else:
            print("Empate en la ronda, no hay puntos.")

    # Jugar todo 
    def jugar(self):
        while self.jugador1.puntos < self.puntaje_maximo and self.jugador2.puntos < self.puntaje_maximo:
            print("\nIniciando nueva mano...")
            self.mazo.mezclar() 
            
            # Repartir 
            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))
            
            # Jugar 
            for _ in range(3):
                self.jugar_ronda()

            # Mostrar 
            print(f"Puntaje: {self.jugador1.nombre}: {self.jugador1.puntos}, {self.jugador2.nombre}: {self.jugador2.puntos}")
        
        # Determinar 
        if self.jugador1.puntos >= self.puntaje_maximo:
            print(f"\n¡{self.jugador1.nombre} ha ganado el juego con {self.jugador1.puntos} puntos!")
        else:
            print(f"\n¡{self.jugador2.nombre} ha ganado el juego con {self.jugador2.puntos} puntos!")

# Ejemplo 
p = Partida(Jugador("Juan"), Jugador("Pedro"))
p.jugar()
```
