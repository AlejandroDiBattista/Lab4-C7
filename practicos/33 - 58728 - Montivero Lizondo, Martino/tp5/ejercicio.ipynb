## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad): 


        self._precio = 0
        self.precio=precio

        self._cantidad = 0
        self.cantidad = cantidad
        self.codigo = codigo
        self.nombre = nombre
        self.tipo=tipo

    @property
    def precio (self):
        return self._precio
    
    
    @precio.setter
    def precio (self,nprecio):
        if nprecio>0: self._precio=nprecio

    @property
    def cantidad(self):
        return self._cantidad
    
    @cantidad.setter
    def cantidad(self,ncantidad):
        if ncantidad > 0: self._cantidad = ncantidad
        

    

    
           
           
    
    
        
        
    def valorTotal(self):
        return self.precio * self.cantidad

        




```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba de la clase Producto

p1 = Producto('0001', 'Coca Cola', 1500, 'gaseosa', 10)

assert p1.codigo == '0001'
assert p1.nombre == 'Coca Cola'
assert p1.precio == 1500

# Calcula el valor total 
assert p1.valorTotal() == 15000 

# Asegura que los valores de precio y cantidad sean validos
p1.precio = -200
assert p1.precio == 1500    # Rechaza el valor negativo

p1.precio = 2000 
assert p1.precio == 2000

p1.cantidad = -5
assert p1.cantidad == 10    # Rechaza el valor negativo

p1.cantidad = 50
assert p1.cantidad == 50
assert p1.valorTotal() == 100000

print("Prueba pasada exitosamente!")
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion = None, codigos=[], tipos=[]):
        self.descripcion = descripcion
        self.codigos = codigos
        self.tipos = tipos

    def esAplicable(self, producto):
        
        return (producto.codigo in self.codigos) or (producto.tipo in self.tipos)



class OfertaDescuento(Oferta):
    def __init__(self, descuento, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.descuento = descuento

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto):return ((producto.precio) * (self.descuento / 100)) * cantidad
        return 0


class Oferta2x1(Oferta):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def calcularDescuento(self, producto, cantidad):
        if self.esAplicable(producto):
            nproductosdes = cantidad // 2
            descuento = nproductosdes * producto.precio
            return descuento
        return 0

```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv

class Catalogo:
    def __init__(self):
        self.lista_productos = []  
        self.lista_ofertas = []    

    @classmethod
    def leer(cls, nombre_archivo):  
        instancia = cls()  
        with open(nombre_archivo, 'r') as archivo:  
            lector = csv.DictReader(archivo)  
            instancia.lista_productos = [
                Producto(row['codigo'], row['nombre'], float(row['precio']), row['tipo'], int(row['cantidad']))
                for row in lector
            ]
        return instancia

    def guardar(self, nombre_archivo):  
        with open(nombre_archivo, 'w', newline='') as archivo:  
            escritor = csv.DictWriter(archivo, fieldnames=['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])  
            escritor.writeheader()
            
            
            for prod in self.lista_productos:  
                escritor.writerow({
                    'codigo': prod.codigo,
                    'nombre': prod.nombre,
                    'precio': prod.precio,
                    'tipo': prod.tipo,
                    'cantidad': prod.cantidad
                })

    def agregar(self, *nuevos_productos):  
        for prod in nuevos_productos:  
            self.lista_productos.append(prod)

    def buscar(self, codigo_busqueda):  
        for prod in self.lista_productos:  
            if prod.codigo == codigo_busqueda:
                return prod
        

    def registrarOferta(self, nueva_oferta):  
        self.lista_ofertas.append(nueva_oferta)

    def buscarOferta(self, prod, cantidad):  
        for oferta in self.lista_ofertas:
            if oferta.esAplicable(prod):
                return oferta
        return None

    def calcularDescuento(self, prod, cantidad):  
        oferta = self.buscarOferta(prod, cantidad)
        if oferta:
            return oferta.calcularDescuento(prod, cantidad)
        return 0

    def vender(self, prod, cantidad):  
        if prod.cantidad >= cantidad:
            prod.cantidad -= cantidad
        

    @property
    def cantidadProductos(self):
        return len(self.lista_productos)

    @property
    def cantidadUnidades(self):
        return sum(prod.cantidad for prod in self.lista_productos)  

    @property
    def valorTotal(self):
        return sum(prod.precio * prod.cantidad for prod in self.lista_productos)  

    def informe(self):
        precio_prom = self.valorTotal / self.cantidadUnidades 
        categorias = {}  
        
        for prod in self.lista_productos:  
            if prod.tipo not in categorias:
                categorias[prod.tipo] = {'unidades': 0, 'precio_total': 0}
            categorias[prod.tipo]['unidades'] += prod.cantidad
            categorias[prod.tipo]['precio_total'] += prod.precio * prod.cantidad
        
        
        informe = f"Catalogo\n"
        informe += f"Cantidad de productos: {self.cantidadProductos}\n"
        informe += f"Cantidad de unidades: {self.cantidadUnidades}\n"
        informe += f"Precio Promedio: $ {precio_prom}\n"
        informe += f"Valor total: $ {self.valorTotal}\n"
        informe += "Tipos de productos: \n"
        
        
        if categorias:
            for tipo, datos in categorias.items():  
                promedio_cat = datos['precio_total'] / datos['unidades'] 
                informe += f" - {tipo} : {datos['unidades']} unidades  {promedio_cat}"
        
        
        informe += "Ofertas:"
        for oferta in self.lista_ofertas:
            if isinstance(oferta, Oferta2x1):
                informe += "Oferta 2x1: {oferta.descripcion}"
           
        
        return informe
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente

- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit): 

        self._nombre = ""
        self.nombre = nombre
        self._cuit = ""
        self.cuit= cuit

    @property
    def nombre (self):
        return self._nombre
    
    @nombre.setter
    def nombre (self,nombrebien):
        
        if nombrebien and nombrebien.strip() and all (not nombrebien.isdigit() for i in nombrebien): 
            self._nombre=nombrebien
            print(f"Cliente: {nombrebien}" )
        
    @property
    def cuit (self):
        return self._cuit
    
    @cuit.setter
    
    def cuit (self,cuitbien):
        
        

        if (len(cuitbien) ==13 and cuitbien[2]=="-" and cuitbien [11]=="-" and cuitbien [:1].isdigit( ) and cuitbien [3:10].isdigit() and cuitbien [12:].isdigit()):
            self._cuit = cuitbien
            print(f"Numero de CUIT: {cuitbien}")

        


```

```python
from datetime import datetime

class Factura:
    numero = 100

    @classmethod
    def ultimaFactura(cls, num):
        cls.numero = num

    @classmethod
    def nuevoNumero(cls):
        cls.numero += 1
        return cls.numero

    def __init__(self, catalogo, cliente):
        self.catalogo = catalogo
        self.cliente = cliente
        self.numFactura = Factura.nuevoNumero()
        self.date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.itemsVendidos = []

    @property
    def cantidadProductos(self):
        productos_unicos = {producto for producto, _ in self.itemsVendidos}
        return len(productos_unicos)

    @property
    def cantidadUnidades(self):
        total_cantidad = sum(cantidad for _, cantidad in self.itemsVendidos)
        return total_cantidad

    @property
    def subtotal(self):
        subtotal_calculado = sum(item.precio * cantidad for item, cantidad in self.itemsVendidos)
        return subtotal_calculado

    @property
    def descuentos(self):
        total_descuentos = 0
        for item, cantidad in self.itemsVendidos:
            total_descuentos += self.catalogo.calcularDescuento(item, cantidad)
        return total_descuentos

    @property
    def total(self):
        total_final = self.subtotal - self.descuentos
        return total_final

    def agregar(self, item_nuevo, cant_nueva):
        if cant_nueva > 0 and item_nuevo.cantidad >= cant_nueva:
            for index, (item, cantidad) in enumerate(self.itemsVendidos):
                if item == item_nuevo:
                    nueva_cantidad = cantidad + cant_nueva
                    if nueva_cantidad <= item_nuevo.cantidad:
                        item_nuevo.cantidad -= cant_nueva
                        self.itemsVendidos[index] = (item_nuevo, nueva_cantidad)
                        return
            self.itemsVendidos.append((item_nuevo, cant_nueva))
            item_nuevo.cantidad -= cant_nueva

    def imprimir(self):
        texto_factura = f"Factura N°: {self.numFactura}\n"
        texto_factura += f"Fecha: {self.date}\n"
        texto_factura += f"Cliente: {self.cliente.nombre} (CUIT: {self.cliente.cuit})\n\n"

        productos_agrupados = {}
        for item, cantidad in self.itemsVendidos:
            if item in productos_agrupados:
                productos_agrupados[item] += cantidad
            else:
                productos_agrupados[item] = cantidad

        for item, cantidad in productos_agrupados.items():
            subtotal_f = item.precio * cantidad
            descuento_f = self.catalogo.calcularDescuento(item, cantidad)
            texto_factura += f"{cantidad} {item.nombre:<30} x ${item.precio:.2f} = ${subtotal_f:.2f}\n"
            
            if descuento_f > 0:
                if item.codigo in ['0001', '0003']:
                    texto_factura += f"Descuento 10% - ${descuento_f:.2f}\n"
                if item.tipo == 'galleta':
                    texto_factura += f"Oferta 2x1 - ${descuento_f:.2f}\n"
        
        texto_factura += f"Subtotal: ${self.subtotal:.2f}\n"
        texto_factura += f"Descuentos: ${self.descuentos:.2f}\n"
        texto_factura += f"TOTAL: ${self.total:.2f}\n"

        return texto_factura
```
