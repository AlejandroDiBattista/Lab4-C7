## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():
    archivoDat = open("datos.dat", "r").readlines()

    ventas = []

    for i in archivoDat:
        fecha = i[0:10].strip()
        producto = i[10:40].strip()
        precio = float(i[40:50].strip())
        cantidad = int(i[50:55].strip())

        v = {
            "Fecha": fecha,
            "Producto": producto,
            "Precio": precio,
            "Cantidad": cantidad
        }

        ventas.append(v)
    return ventas

datos = cargar_datos()

for v in datos:
    print(v)
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales():
    totalImporte = 0
    totalCantidad = 0

    archivoDat = open("datos.dat", "r").readlines()

    ventas = []

    for i in archivoDat:
        fecha = i[0:10].strip()
        producto = i[10:40].strip()
        precio = float(i[40:50].strip())
        cantidad = int(i[50:55].strip())

        v = {
            "Fecha": fecha,
            "Producto": producto,
            "Precio": precio,
            "Cantidad": cantidad
        }

        ventas.append(v)
    
    for j in ventas:
        precio = j["Precio"]
        cantidad = j["Cantidad"]

        totalImporte = totalImporte + (precio * cantidad)
        totalCantidad = totalCantidad + cantidad
    return totalImporte, totalCantidad

importe, cantidad = calcular_totales()
print(f"Las ventas fueron de ${importe} en {cantidad} unidades") 
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas():
    archivoDat = open("datos.dat", "r").readlines()

    ventas = []

    for i in archivoDat:
        fecha = i[0:10].strip()
        producto = i[10:40].strip()
        precio = float(i[40:50].strip())
        cantidad = int(i[50:55].strip())

        v = {
            "Fecha": fecha,
            "Producto": producto,
            "Precio": precio,
            "Cantidad": cantidad
        }

        ventas.append(v)

    dicVentas = {}

    for v in ventas:
        prod = v["Producto"]
        cant = v["Cantidad"]

        if prod in dicVentas:
            dicVentas[prod] = dicVentas[prod] + cant
        else:
            dicVentas[prod] = cant

    return dicVentas

ventas = unidades_vendidas()


def listar_ventas(ventas):
    print("Unidades Vendidas por Producto\n")
    
    for i, j in ventas.items():
        print(f"{i}: {j} unidades")

listar_ventas(ventas)
````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def precio_promedio():
    archivoDat = open("datos.dat", "r").readlines()

    ventas = []

    for i in archivoDat:
        fecha = i[0:10].strip()
        producto = i[10:40].strip()
        precio = float(i[40:50].strip())
        cantidad = int(i[50:55].strip())

        v = {
            "Fecha": fecha,
            "Producto": producto,
            "Precio": precio,
            "Cantidad": cantidad
        }

        ventas.append(v)

    dicPromedios = {}
    dicCont = {}

    for j in ventas:
        prod = j["Producto"]
        pre = j["Precio"]

        if prod not in dicPromedios:
            dicPromedios[prod] = 0
            dicCont[prod] = 0

        dicPromedios[prod] = dicPromedios[prod] + pre
        dicCont[prod] = dicCont[prod] + 1

    for l in dicPromedios:
        promedio = dicPromedios[l] / dicCont[l]
        dicPromedios[l] = promedio

    return dicPromedios

precios = precio_promedio()


def listar_precios(precios):
    for i, j in precios.items():
        print(f"Producto: {i}, Precio Promedio: ${j:.2f}")

listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas():
    archivoDat = open("datos.dat", "r").readlines()

    ventas = []

    for i in archivoDat:
        fecha = i[0:10].strip()
        producto = i[10:40].strip()
        precio = float(i[40:50].strip())
        cantidad = int(i[50:55].strip())

        v = {
            "Fecha": fecha,
            "Producto": producto,
            "Precio": precio,
            "Cantidad": cantidad
        }

        existe = False

        for j in ventas:
            if j["Producto"] == producto:
                j["Cantidad"] = j["Cantidad"] + cantidad
                j["Importe Total"] = j["Importe Total"] + (precio * cantidad)
                
                existe = True
                break

        if not existe:
            ventas.append({
                "Producto": producto,
                "Cantidad": cantidad,
                "Importe Total": precio * cantidad
            })
    
    for l in ventas:
        l["Precio Promedio"] = l["Importe Total"] / l["Cantidad"] if l["Cantidad"] > 0 else 0

    return ventas

resumen = resumen_ventas()


def informe_ventas(resumen):
    resumen.sort(key=lambda x: x["Producto"])

    print("                          Informe de Ventas")
    print("-" * 75)
    print(f"{'Producto':<15} {'Precio Promedio':<20} {'Cantidad Vendida':<20} {'Importe Total':<15}")
    print("-" * 75)

    for i in resumen:
        print(f"{i['Producto']:<18} {i['Precio Promedio']:<24.2f} {i['Cantidad']:<15} {i['Importe Total']:<15.2f}")

informe_ventas(resumen)
````
