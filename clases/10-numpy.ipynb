```python
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 20)
# print(x)
y = 1*x**3 -4 *x**2 + 2 *x + 1
np.random.seed(0)
yr = y + np.random.randn(20)*200
plt.scatter(x, yr)
plt.plot(x, y, 'r', linewidth=20)
coeficientes = np.polyfit(x, yr, 4)
polinomio = np.poly1d(coeficientes)
plt.plot(x, polinomio(x), 'g', linewidth=10)
print(coeficientes)
```

```python
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt 
import os
from scipy.ndimage import sobel

# Cargar la imagen
image = Image.open('../../../señal.png')

# Convertir la imagen a un array de numpy
im = np.array(image)

print(im.shape)
# im[0:100:,:,2] = 0
# Mostrar los 3 canales rojo, verde, azul uno al lado del otro
fig, (r,g,b,a) = plt.subplots(1, 4, figsize=(15, 5))

r.imshow(im[:, :, 0], cmap='Reds')
r.set_title('Canal Rojo')

g.imshow(im[:, :, 1], cmap='Greens')
g.set_title('Canal Verde')

b.imshow(im[:, :, 2], cmap='Blues')
b.set_title('Canal Azul')

a.imshow(im[:, :, 3], cmap='Grays')
a.set_title('Canal Alfa')


plt.show()

```
